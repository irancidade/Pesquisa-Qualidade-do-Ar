```python
import pandas as pd
acolunas= ['segundo','CO2','CO','PM10','PM25','ConcPM1','ConcPM25','Pressao','Temp','Umidade'] 

dadosF = pd.read_csv("Fuzzydirigivel.csv",sep=",",names=acolunas)
```

```python
import numpy as np
dadosF.describe().T
```

```python
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt
x=dadosF['PM25']
y=dadosF['Pressao']
plt.scatter(x,y, s=4, marker='o', c='red')
plt.title('Dispersão de CO') #adicionando o título
plt.xlabel('partículas em µg/m³') #definindo nome do eixo X
plt.ylabel('Pressao em Pascal') #definindo nome do eixo Y
plt.gca().invert_yaxis() # Invertendo o eixo y
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Supondo que 'dadosT' seja seu DataFrame
x = dadosF['Temp']
y = dadosF['CO']
z = dadosF['Pressao']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z,s=2, c='red', marker='o')

ax.set_title('CO x Pressao x Temperatura')
ax.set_xlabel('Temperatura em ºC')
ax.set_ylabel('CO em ppm')
ax.set_zlabel('Pressâo')

ax.invert_zaxis()  # Invertendo o eixo Z

plt.show()

```
